```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "A": [125, 210, 345, 480, 512, 678, 890, 745, 932, 600,
         415, 299, 850, 720, 530, 410, 999, 860, 275, 660]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
print("Mean:")
print(df.mean())

print("Median:")
print(df.median())

print("Standard Deviation:")
print(df.std())

print("Variance:")
print(df.var())

print("Skewness:")
print(df.skew())

print("Kurtosis:")
print(df.kurt())
```

```python
sk = df['A'].skew()
plt.figure(figsize=(8, 5))
sns.histplot(df['A'], kde=True, bins=5, color="skyblue")
plt.title(f"Skewness Plot\nSkewness = {sk:.2f}", fontsize=14)
plt.xlabel("Value")
plt.ylabel("Count")
plt.show()
```

```python
kt = df['A'].kurt()
plt.figure(figsize=(8, 5))
sns.histplot(df['A'], kde=True, bins=5, color="lightcoral")
plt.title(f"Kurtosis Plot\nKurtosis = {kt:.2f}", fontsize=14)
plt.xlabel("Value")
plt.ylabel("Count")
plt.show()
```
