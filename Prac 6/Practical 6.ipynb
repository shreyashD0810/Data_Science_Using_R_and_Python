```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('tips')
```

```python
data
```

```python
sns.histplot(data['total_bill'], kde=True, color='skyblue')
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.countplot(x='day', data=data, palette='pastel')
plt.title("Number of Records by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x='sex', y='tip', data=data, palette='Set2')
plt.title("Tip Amount by Gender")
plt.xlabel("Sex")
plt.ylabel("Tip")
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=data)
plt.title("Tip vs Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
